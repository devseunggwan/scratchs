```python
import os

import httpx
from dotenv import load_dotenv

load_dotenv()
```

```python
X_API_KEY = os.getenv("X_API_KEY")
CHAIN = "klaytn"
COLLECTION_ID = "0x50f718ca7dc27393dbcc2b41c5867c588e2b48fb"

HEADERS = {
    "x-api-key": X_API_KEY,
    "Content-Type": "application/json",
}
```

```python
url_get_contract = (
    f"https://api.opensea.io/api/v2/chain/{CHAIN}/contract/{COLLECTION_ID}"
)
```

```python
response = httpx.get(url_get_contract, headers=HEADERS)
response = response.json()

collection_slug = response["collection"]
```

```python
url_get_nfts = f"https://api.opensea.io/api/v2/collection/{collection_slug}/nfts"
```

```python
response = httpx.get(url_get_nfts, headers=HEADERS)
response = response.json()
```
