```python
import requests
```

```python
API_KEY = "" # Enter your API key here
```

```python
# Collections
# url: https://api.reservoir.tools/collections/v7
# docs: https://docs.reservoir.tools/reference/getcollectionsv7

url = "https://api.reservoir.tools/collections/v7"
params = {
    "api_key": API_KEY,
    "limit": 1000,
    "sortBy": "updatedAt",
}
```

```python
resp = None
collections = []

while True:
    if resp is not None:
        params["continuation"] = resp.get("continuation")
    
    
    resp = requests.get(url=url, params=params)    
    resp = resp.json()
    collections.extend(resp.get("collections"))
    
    if resp.get("continuation") in [None, ""]:
        break
```
