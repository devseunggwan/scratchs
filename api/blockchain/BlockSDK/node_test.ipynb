```python
import os

import pandas as pd
import httpx
from web3 import Web3
from dotenv import load_dotenv
load_dotenv()
```

```python
API_TOKEN = os.getenv("API_TOKEN")
NODE_RPC_KEY = os.getenv("NODE_RPC_KEY")

VERSION = "v3"
CHAIN = "eth"
URL = f"https://testnet-api.blocksdk.com/{VERSION}/{CHAIN}"

web3 = Web3(Web3.HTTPProvider(f"https://main-eth-rpc.blocksdk.com/{NODE_RPC_KEY}"))
```

## API TEST

### 현재 블록 조회

```python
endpoint = "info"

endpoint = os.path.join(URL, endpoint)

resp = httpx.get(endpoint, params={"api_token": API_TOKEN})
resp.json()
```

### 블록 정보 조회

```python
block_number = 10000000
endpoint = f"block/{block_number}"

params = {
    "api_token": API_TOKEN,
    "offset": 0,
    "limit": 1000
}

resp = httpx.get(os.path.join(URL, endpoint), params=params)
resp.json()
```
