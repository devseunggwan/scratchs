```python
from datetime import datetime

import pandas as pd
pd.options.display.float_format = '{:.5f}'.format
```

```python
df = pd.read_csv('/workspaces/scratchs/data-processing/202301023/tb_cryptocurrency_202310231739.csv', dtype=object)
df
```

```python
df["date"] = df["date"].apply(str)
df["time"] = df["time"].apply(str)
df["timestamp"] = df["date"] + df["time"]
df
```

```python
del df["date"], df["time"]
df
```

```python
df["timestamp"] = df["timestamp"].apply(lambda x: datetime.strptime(x, "%Y%m%d%H%M%S"))
df["timestamp"] = df["timestamp"].apply(lambda x: datetime.timestamp(x))
df["timestamp"] = df["timestamp"].apply(int)
df
```

```python
df.to_csv('/workspaces/scratchs/data-processing/202301023/tb_cryptocurrency_202310231739_result.csv', index=False)
```
