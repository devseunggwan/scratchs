```python
import os
import json

import aiohttp
from dotenv import load_dotenv

load_dotenv()
```

```python
url = os.getenv('API_URL')
headers = {
    "Authorization": os.getenv('API_KEY')
}

url_collections = os.getenv('API_COLLECTIONS')
url_collection_info = os.getenv('API_COLLECTION_INFO')
url_nft_list = os.getenv('API_NFT_LIST')
```

```python
def write_json(data, filename='data.json'):
    with open(filename, 'w', encoding="utf-8") as f:
        json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
async with aiohttp.ClientSession() as session:
    async with session.get(os.path.join(url, url_collections), headers=headers) as response:
        data = await response.json()
        
        write_json(data, './data/collections.json')
        
        
```

```python
async with aiohttp.ClientSession() as session:
    for collection in data["result"]["collections"]:
        params = {
            "address": collection["address"],
            "offset": 0,
            "limit": 1000
        }
        
        async with session.get(os.path.join(url, url_nft_list), params=params, headers=headers) as response:
            data = await response.json()
            
            write_json(data, f'./data/nft_list_{collection["address"]}.json')
```

```python
async with aiohttp.ClientSession() as session:
    for collection in data["result"]["collections"]:
        params = {
            "address": collection["address"]
        }
        
        async with session.get(os.path.join(url, url_collection_info), params=params, headers=headers) as response:
            data = await response.json()
            
            write_json(data, f'./data/collection_info/{collection["address"]}.json')
```
