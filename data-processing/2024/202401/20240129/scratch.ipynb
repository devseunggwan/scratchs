```python
import pandas as pd
```

```python
df = pd.read_csv("./data/transaction_table.csv")
df
```

```python
df.sort_values(by="TIMESTAMP", ascending=True, inplace=True)
df
```

```python
"""
    {
        "{collection_id}": {
            "{token_id}": {
                "block_number": value,
                "amount": value,
            }
        }, 
        ...
    }

"""

ACCOUNT = "0x1c593c369136264Abfb908b4B20e74c9a5949417"
my_account = {}


for row in df.loc[df["NETWORK"] == "ethereum"].to_dict(orient="records"):
    if row["FROM"] == ACCOUNT:
        if row["COLLECTION_ID"] not in my_account:
            my_account[row["COLLECTION_ID"]] = {}

        if row["TOKEN_ID"] not in my_account[row["COLLECTION_ID"]]:
            my_account[row["COLLECTION_ID"]][row["TOKEN_ID"]] = -row["AMOUNT"]
        else:
            my_account[row["COLLECTION_ID"]][row["TOKEN_ID"]] -= row["AMOUNT"]

    if row["TO"] == ACCOUNT:
        if row["COLLECTION_ID"] not in my_account:
            my_account[row["COLLECTION_ID"]] = {}

        if row["TOKEN_ID"] not in my_account[row["COLLECTION_ID"]]:
            my_account[row["COLLECTION_ID"]][row["TOKEN_ID"]] = row["AMOUNT"]
        else:
            my_account[row["COLLECTION_ID"]][row["TOKEN_ID"]] += row["AMOUNT"]
```
