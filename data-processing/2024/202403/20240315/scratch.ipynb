```python
import json
from decimal import Decimal

import httpx
```

```python
with open("./data/sample.json") as f:
    samples = json.load(f)

samples
```

```python
sales = samples["sales"]
sales
```

```python
len([sale for sale in sales if sale["orderSide"] == "bid"])
```

```python
[sale for sale in sales if sale["orderSide"] == "bid"]
```

```python
nfts = []

for item in [sale for sale in sales if sale["orderSide"] == "bid"]:
    nft = {
        "network": "ethereum",
        "collectionId": item["token"]["collection"]["id"],
        "tokenId": item["token"]["tokenId"],
        "imgUrl": item["token"]["image"],
        "price": Decimal(item["price"]["amount"]["decimal"]),
        "currency": item["price"]["currency"]["symbol"],
        "priceUsd": round(Decimal(item["price"]["amount"]["usd"]), 4),
        "priceKrw": None,
        "amount": int(item["amount"]),
        "market": item["fillSource"],
        "fromAccount": item["from"],
        "toAccount": item["to"],
        "timestamp": item["timestamp"],
        "txnLink": item["txHash"],
        "collectionName": item["token"]["collection"]["name"],
        "nftName": item["token"]["name"],
    }

    nfts.append(nft)
```
