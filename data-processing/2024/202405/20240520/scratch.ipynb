```python
import json

from dotenv import load_dotenv
from tqdm import tqdm

from service import NftCurationLLM

load_dotenv()
```

```python
def read_json(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data

def write_json(data, filename):
    with open(filename, 'w') as f:
        json.dump(data, f, indent=4, ensure_ascii=False)
        
```

```python
top100 = read_json('top100.json')
top100 = tqdm(top100["collections"])
```

```python
NETWORK = "ethereum"

llm = NftCurationLLM()

for idx, item in enumerate(top100):
    top100.set_description(f'Processing {idx}: {item["name"]}')
    
    try:
        result = llm(network=NETWORK, collection_id=item["id"])
        write_json(result, f'{NETWORK}/{NETWORK}_{item["id"]}_{result["created_date"]}.json')
    except Exception as e:
        print(f'Error processing {item["id"]}: {e}')
```
