```python
import os
import re

import httpx
from dotenv import load_dotenv

load_dotenv()
```

```python
headers = {
    "x-api-key": os.getenv("RESERVOIR_API_KEY"),
}
url = "https://api-polygon.reservoir.tools/tokens/v7"

params = {
    "collection": "0x3b728e0c366fb6d6e79e1c7942f9e4107f2b2eb8",
    "sortBy": "updatedAt",
    "limit": 1000,
}
```

```python
resp = httpx.get(url, headers=headers, params=params)
resp.json()
```

```python
nft_images = [
    nft["token"]["imageSmall"]
    for nft in resp.json()["tokens"]
    if "imageSmall" in nft["token"] and nft["token"]["imageSmall"] is not None
]
len(nft_images)
```

```python
def filter_image_url(image_urls: list[str]) -> list[str]:
    return [
        image_url
        for image_url in image_urls
        if re.match(
            r".*(png|jpeg|gif|webp).*",
            image_url,
            re.IGNORECASE,
        )
    ]
```

```python
nft_images = filter_image_url(nft_images)
len(nft_images)
```

### GPT에서 사용할 수 있는 데이터 필터링

```python
from PIL import Image

filtered_images = []

for image_url in nft_images:
    image = Image.open(httpx.get(image_url))

    width, height = image.size

    if width == height:
        filtered_images.append(image_url)
```

```python
len(filtered_images)
```

```python
with open("nft_images.txt", "w") as f:
    f.write("\n".join(filtered_images))
```
