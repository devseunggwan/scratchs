```python
import pandas as pd

df = pd.read_csv("data/data_exported_2025-12-01_014832.csv")
```

```python
df_metrics = df[df["body"].str.contains(r"\"\d{1,4}\"", regex=True)][
    ["timestamp", "body"]
]
df_metrics
```

```python
df_metrics["timestamp"] = pd.to_datetime(df_metrics["timestamp"]) + pd.Timedelta(
    hours=9
)
df_metrics["body"] = df_metrics["body"].str.replace('"', "").astype(int)
# df_metrics.to_csv("data/data_exported_2025-12-01_014832_prep.csv", index=False)
```

```python
df_metrics
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 시간(hour)별 평균 계산
sns.set_style("whitegrid")

df_metrics["hour"] = df_metrics["timestamp"].dt.hour
df_plot = df_metrics.groupby("hour")["body"].mean().reset_index()

plt.figure(figsize=(20, 10))

# 00~08시인 경우를 위한 색상 처리
bar_colors = ["red" if 0 <= hour <= 8 else "C0" for hour in df_plot["hour"]]

barplot = sns.barplot(
    x="hour",
    y="body",
    data=df_plot,
    hue="hour",
    palette=dict(zip(df_plot["hour"], bar_colors)),
    legend=False,
)

plt.xlabel("Hour")
plt.ylabel("Average Metric Value")
plt.xticks(range(24), [f"{h}" for h in range(24)])

# bar_label 사용하여 정확하게 바 위에 값 표시
for container in barplot.containers:
    barplot.bar_label(container, fmt="%.0f", fontsize=9)

plt.tight_layout()
plt.show()
```
