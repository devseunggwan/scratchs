```python
import gzip
import orjson
from pathlib import Path


src = Path("data/sample.json.gz")
dst = Path("data/sample_copy.txt")


with gzip.open(src, "rt", encoding="utf-8") as fin:
    data = fin.readlines()
    data = [item.strip() for item in data]
    data = [orjson.loads(item) for item in data if item != ""]
```

```python
margin = 10
trigger = False

for idx, item in enumerate(data):
    if item["0"] == 20:
        print(item)
        print(idx)
        trigger = True

    if trigger:
        print(item)
        print(idx)
        margin -= 1

    if margin == 0:
        trigger = False
        margin = 10
```
