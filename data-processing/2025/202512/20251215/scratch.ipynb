```python
import duckdb

df = duckdb.read_json("data/2025-12-15.json")
```

```python
query = """
    WITH st0 AS (
        SELECT UNNEST(data.lastChangeStatuses) AS items
          FROM df
    )
    SELECT items.orderId,
           items.productOrderId,
           items.lastChangedType,
           items.paymentDate,
           items.lastChangedDate,
           items.productOrderStatus,
           items.receiverAddressChanged
      FROM st0
"""

prep_df = duckdb.query(query)
```

```python
query = """
    SELECT productOrderId
         , lastChangedDate
      FROM prep_df
     WHERE 1 = 1
       AND productOrderId::INTEGER >= 2025121400000000
       AND productOrderId::INTEGER <  2025121500000000
"""

duckdb.query(query)
```

```python
query = """
    WITH st0 AS (
        SELECT content
        FROM read_json('data/0.json.gz')
    )
    SELECT UNNEST(st0.content) AS items
      FROM st0
"""

kakao_shopping_df = duckdb.query(query=query)
kakao_shopping_df.shape
```
