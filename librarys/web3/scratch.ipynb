```python
import json
import os

from web3 import Web3, HTTPProvider
from dotenv import load_dotenv
load_dotenv()
```

```python
def write_json(data, filename="data.json"):
    with open(filename, "w") as f:
        json.dump(data, f, ensure_ascii=False)
```

```python
# URL = os.getenv("KLAYTN_URL")
URL = "https://rpc.mevblocker.io/fast"
```

```python
web3 = Web3(HTTPProvider(URL))
print(web3.is_connected())
```

```python
web3.eth.get_block("latest", True)
```

```python
def transfer_transaction(transaction):
    transaction = transaction.__dict__
    transaction["blockHash"] = transaction["blockHash"].hex()
    transaction["hash"] = transaction["hash"].hex()
    transaction["r"] = transaction["r"].hex()
    transaction["s"] = transaction["s"].hex()
        
    return transaction
```

```python
sample = web3.eth.get_block("latest", True)
sample = sample.__dict__
sample["extraData"] = sample["extraData"].hex()
sample["hash"] = sample["hash"].hex()
sample["logsBloom"] = sample["logsBloom"].hex()
sample["mixHash"] = sample["mixHash"].hex()
sample["nonce"] = sample["nonce"].hex()
sample["parentHash"] = sample["parentHash"].hex()
sample["receiptsRoot"] = sample["receiptsRoot"].hex()
sample["sha3Uncles"] = sample["sha3Uncles"].hex()
sample["stateRoot"] = sample["stateRoot"].hex()
sample["transactionsRoot"] = sample["transactionsRoot"].hex()
sample["transactions"] = [transfer_transaction(transaction) for transaction in sample["transactions"]]

sample
```

```python
write_json(sample, "sample.json")
```
