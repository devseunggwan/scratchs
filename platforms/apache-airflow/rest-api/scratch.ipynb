```python
import os

import requests
from requests.auth import HTTPBasicAuth
from dotenv import load_dotenv

load_dotenv()

AIRFLOW_URL = os.getenv("AIRFLOW_URL")
DAG_ID = os.getenv("DAG_ID")

headers = {
    "Accept": "application/json",
    "Content-Type": "application/json",
}
auth = HTTPBasicAuth(os.getenv("AIRFLOW_USER"), os.getenv("AIRFLOW_PASSWORD"))
```

```python
dag_runs_url = f"{AIRFLOW_URL}/dags/{DAG_ID}/dagRuns"
params = {
    "limit": 100,
    "order_by": "-execution_date",
}

resp = requests.get(dag_runs_url, params=params, headers=headers, auth=auth)

dag_runs = resp.json()
dag_runs
```

```python
dag_run_id = dag_runs["dag_runs"][0]["dag_run_id"]
tasks_url = f"{AIRFLOW_URL}/dags/{DAG_ID}/dagRuns/{dag_run_id}/taskInstances"

resp = requests.get(tasks_url, headers=headers, auth=auth)
resp.raise_for_status()

tasks = resp.json()
tasks
```
